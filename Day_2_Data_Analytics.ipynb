## **Central Limit Theorem and Sampling Methods**

Importing the necessary libraries and loading the IRIS dataset

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

Iris = sns.load_dataset('iris')
print(Iris.head())
```

Creating a random sample

```python
Random_sample =Iris.sample(n=30, random_state=30) #To create 30 sample and to generate the same randomisation when re-executed
print(Random_sample)
```

Creating sample means for the Sample data generated

```python
sample_means = []
for i in range(100):
  sample = Iris.sample(n=30, random_state=i)  # Generate a new sample each time
  sample_mean = sample['sepal_length'].mean()
  sample_means.append(sample_mean)  # Append mean to the sample_means list
print(sample_means)
```

Ploting Distribution of Sample Means


```python
plt.figure(figsize=(8, 5))
plt.hist(sample_means, bins=15, color='green', edgecolor='black')
plt.title("Distribution of Sample Means for Sepal Length)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

 Systematic Sampling: Taking 20% of the dataset

```python
import random
step_size = len(Iris) // int(0.2 * len(Iris))
start = random.randint(0, step_size - 1)
systematic_sample = Iris[start::step_size] #To Sample the sataset from a random start index
print(systematic_sample)
```

Plotting the Original vs Sample Dataset

```python
plt.figure(figsize=(10, 6))
plt.scatter(range(len(Iris)), Iris['sepal_length'], color='yellow', alpha=0.5, label='Original Dataset')
plt.scatter(systematic_sample.index, systematic_sample['sepal_length'], color='red', label='Systematic Sample')
plt.title("Systematic Sampling: Sepal Length")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.legend()
plt.show()
```
