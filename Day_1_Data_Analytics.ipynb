Importing the Dataset and libraries. The dataset is taken from Kagle


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import mode

from google.colab import files
uploaded = files.upload()
```

Printing the inserted dataset

```python
import io

df = pd.read_excel(io.BytesIO(uploaded['HealthCare.xlsx']))
df.head()
```

Statistical Analysis for the column Cost_Per_Day

```python
Mean = df['Cost_Per_Day'].mean()
Median = df['Cost_Per_Day'].median()
Mode = mode(df['Cost_Per_Day'])
SD = df['Cost_Per_Day'].std()
One_SD_minus = Mean - SD
One_SD_plus = Mean + SD

print(f"Statistics for Cost_Per_Day:")
print(f"Mean: {Mean}")
print(f"Median: {Median}")
print(f"Mode: {Mode}")
print(f"Standard Deviation: {SD}")
print(f"1-SD: {One_SD_minus}")
print(f"1+SD: {One_SD_plus}")
```

Plotting the graphs for these calculations

```python
#Plotting Length_of_stay vs Cost_per_day for cost analysis
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Length_Of_Stay', y='Cost_Per_Day')
plt.axhline(Mean, color='red', linestyle='--', label='Mean')
plt.axhline(Median, color='green', linestyle='--', label='Median')
plt.axhline(Mode[0], color='purple', linestyle='--', label='mode') #Using Mode[0] here to only take single value to plot the line
plt.axhline(One_SD_minus, color='orange', linestyle='--', label='1-SD')
plt.axhline(One_SD_plus, color='blue', linestyle='--', label='1+SD')
plt.title('Length Of Stay vs Cost Per Day')
plt.xlabel('Length_Of_Stay')
plt.ylabel('Cost_Per_Day')
plt.legend()
plt.show()


#Plotting Count of pattients to gender for patient demographics
plt.figure(figsize=(10, 6))
plt.bar(df['Gender'].value_counts().index, df['Gender'].value_counts().values) #taking the count seperately and the entries separately
plt.title('Count of Patients by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
